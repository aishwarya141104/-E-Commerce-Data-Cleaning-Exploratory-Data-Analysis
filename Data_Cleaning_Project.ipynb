```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month_name()
df["Day"] = df["Date"].dt.day_name()
```

```python
df["Product"].value_counts().plot(kind="bar")

plt.title("Orders by Product")
plt.xlabel("Product")
plt.ylabel("Orders")
plt.show()
```

```python
df["PaymentMethod"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")
plt.title("Payment Methods")
plt.show()
```

```python
df["OrderStatus"].value_counts().plot(kind="bar")

plt.title("Order Status")
plt.show()
```

```python
monthly = df.groupby("Month")["TotalPrice"].sum()

monthly.plot(marker="o")

plt.title("Monthly Sales")
plt.show()
```

```python
df.groupby("Product")["TotalPrice"].sum().sort_values().plot(kind="barh")

plt.title("Revenue by Product")
plt.show()
```

```python
df["ReferralSource"].value_counts().plot(kind="bar")

plt.title("Referral Sources")
plt.show()
```

```python
df["CouponCode"].value_counts().plot(kind="bar")

plt.title("Coupon Usage")
plt.show()
```

```python
plt.hist(df["Quantity"], bins=5)

plt.title("Quantity Distribution")
plt.show()
```

```python
plt.hist(df["UnitPrice"], bins=20)

plt.title("Unit Price Distribution")
plt.show()
```

```python
df.groupby("Year")["TotalPrice"].sum().plot(kind="bar")

plt.title("Sales by Year")
plt.show()
```

```python
df["ItemsInCart"].value_counts().sort_index().plot()

plt.title("Items in Cart")
plt.show()
```

```python
df.groupby("CustomerID")["TotalPrice"].sum().nlargest(10).plot(kind="bar")

plt.title("Top Customers")
plt.show()
```

```python
df.groupby("Product")["UnitPrice"].mean().plot(kind="bar")

plt.title("Average Product Price")
plt.show()
```

```python
df["Day"].value_counts().plot(kind="bar")

plt.title("Orders by Day")
plt.show()
```

```python
corr = df[["Quantity","UnitPrice","ItemsInCart","TotalPrice"]].corr()

plt.imshow(corr)

plt.xticks(range(4), corr.columns)
plt.yticks(range(4), corr.columns)

plt.colorbar()

plt.show()
```
